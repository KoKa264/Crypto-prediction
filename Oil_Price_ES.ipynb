```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
data=pd.read_csv('Brent_Oil_Prices_2018_2022.csv',index_col="Date")
data
```

```python
data.plot(figsize=(15,10))
plt.show()
```

```python
model = ExponentialSmoothing(endog = data.Price,
                           trend="add",
                           seasonal="add",
                           seasonal_periods=7).fit()
```

```python
predictions=model.forecast(steps=100)
```

```python
data['Price']['2018-01-02':].plot(figsize=(20,10))
predictions.plot()
```
